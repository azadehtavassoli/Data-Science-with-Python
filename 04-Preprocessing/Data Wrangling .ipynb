# Concatenating


```python
my_source1 = pd.DataFrame([[0,1,13,17,17],
[0,2,14,14,15],
[0,3,17,12,20],
[0,4,12,18,19]])
my_source1[0] = ['babak', 'raha', 'sara', 'reza']
my_source1
```

```python
my_source2 = pd.DataFrame([[0,1,13, 17],
[0,2,15, 20],
[0,3, 14, 19],
[0,4, 20, 19],
[0,5, 15, 18],
[0,6, 14, 12]])
my_source2[0]=['babak', 'baran', 'sara', 'arash', 'mahan', 'reza'] 
my_source2
```

```python
# axis=0 inedex and axi0=1 col
my_concat = pd.concat([my_source1, my_source2], axis=1, ignore_index=True)
my_concat 
```

```python
my_concat = pd.concat([my_source1, my_source2], axis=0, ignore_index=True)
my_concat 
```

```python
#del col 4
my_concat.drop([4], axis=1, inplace=True)
my_concat
```

```python
# del duplicates
my_concat.drop_duplicates(inplace=True)
my_concat
```

```python
# because  index is not correct
my_concat.reset_index(drop=True, inplace=True)
my_concat
```

```python
smartphones.rename(index=smartphones.Name, inplace=True)
smartphones.drop(['Name', 'Company'], axis=1, inplace=True)
smartphones
```

# Normalize Data

```python
#standardization
from sklearn.preprocessing import scale, normalize, minmax_scale
```

```python
scale_data = minmax_scale(smartphones_data, feature_range=(0, 100))
scale_data
```

```python
df_data = pd.DataFrame(scale_data,
index=smartphones_data.index, columns=smartphones_data.columns)
df_data
```

```python
scale_data = minmax_scale(smartphones_data, feature_range=(0,1))
df_data = pd.DataFrame(scale_data,
index=smartphones_data.index, columns=smartphones_data.columns)
df_data

```
