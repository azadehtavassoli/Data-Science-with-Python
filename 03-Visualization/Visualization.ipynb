# matplotlib

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Iran pop from 1960 to 2016 
years = [1960, 1970, 1980, 1990, 2000, 2010, 2016]
iran_pop = [21.91, 28.51, 38.67, 56.23, 66.13, 74.57, 80.28]
```

# Line Plot

```python
plt.plot(years, iran_pop)
plt.show()
```

# Scatter Plot

```python
plt.scatter(years, iran_pop)
plt.show()
```

# Customize Scatter plot

```python
plt.scatter(np.arange(6),city_pop)
plt.xticks([0, 1, 2, 3, 4, 5],['Tehran', 'Mashhad', 'Isfahan', 'Karaj', 'Tabriz', 'Shiraz'],rotation=60,fontsize='large')
plt.yticks([1000000, 2000000, 7000000],['1M','2M','7M'])
plt.title('Iran\'s most populated city')
plt.show()                          
```

```python
#s size, c=color,
pop_size = np.array([7153309, 2307177, 1547164, 1448075, 1424641, 1249942])/30000
Colors = ['violet', 'green', 'orange', 'tomato', 'blue', 'pink']
plt.scatter(np.arange(6),city_pop, s=pop_size,c=Colors, alpha=1)
#margin is spce for each side
plt.margins(0.1)
plt.xticks([0, 1, 2, 3, 4, 5],['Tehran', 'Mashhad', 'Isfahan', 'Karaj', 'Tabriz', 'Shiraz'],rotation=60,fontsize='large')
plt.yticks([1000000, 2000000, 7000000],['1M','2M','7M'])
plt.title('Iran\'s most populated city')
plt.text(0,7153309,'iran Capital',fontsize=15)
plt.ylim([0,8000000])
#address text on plot
plt.text(0,7153309,'iran Capital',fontsize=15)
plt.show()
```

# Multiple Plot

```python
turkey_pop=[27.47,34.88,43.98,53.92,63.24,72.33,79.51]
plt.plot(years, iran_pop,ls='-',lw=2,marker='+',mew=8)
plt.plot(years, turkey_pop,ls='--',lw=1)
plt.show()
    
    
```

```python
# add x,y lable, title , legend
turkey_pop=[27.47,34.88,43.98,53.92,63.24,72.33,79.51]
plt.plot(years, iran_pop,ls='-',lw=2,marker='+',mew=8)
plt.plot(years, turkey_pop,ls='--',lw=1)
plt.title('Iran vs Turkey Population')
plt.xlabel('Year')
plt.ylabel('Population')
plt.legend(['iran','Turkey'],loc='best')
plt.show()
```

```python
# add yticks and grid
turkey_pop=[27.47,34.88,43.98,53.92,63.24,72.33,79.51]
plt.plot(years, iran_pop,ls='-',lw=2,marker='+',mew=8)
plt.plot(years, turkey_pop,ls='--',lw=1)
plt.yticks([20,30,40,50,60,70,80],['20M', '30M', '40M', '50M' ,'60M', '70M', '80M',])
plt.title('Iran vs Turkey Population')
plt.grid()
plt.xlabel('Year')
plt.ylabel('Population')
plt.legend(['iran','Turkey'],loc='best')
plt.show()
```

```python
#add arrow, annotate
turkey_pop=[27.47,34.88,43.98,53.92,63.24,72.33,79.51]
plt.plot(years, iran_pop,ls='-',lw=2,marker='+',mew=8)
plt.plot(years, turkey_pop,ls='--',lw=1)
plt.yticks([20,30,40,50,60,70,80],['20M', '30M', '40M', '50M' ,'60M', '70M', '80M',])
plt.title('Iran vs Turkey Population')
plt.grid()
plt.xlabel('Year')
plt.ylabel('Population')
plt.legend(['iran','Turkey'],loc='best')
#add arrow, annotate
plt.annotate('IranIraq War',xy=(1980,38.67) ,xytext=(1990,34.88) ,arrowprops=dict(facecolor='silver',width=4),fontsize=15)
plt.show()
```

# Subplot

# Seaborn


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
import pandas as pd
smartphones = pd.read_csv("C:\dataset-F\smartphones.csv")
smartphones
```

# Strip Plot


```python
sb.stripplot(y='Capacity',data=smartphones,size=10)
plt.show()
```

```python
print(smartphones)
sb.stripplot(x='OS',y='Capacity',data=smartphones,size=10, jitter=True)
#sb.swarmplot(x='OS', y='Capacity', data=smartphones)
plt.show()
```

# Swarm Plot

```python
# This function is similar to stripplot(), but the points are adjusted (only along the categorical axis)
#so that they dont overlap
#hue show color
plt.figure(figsize=(8,5))
sb.swarmplot(x='OS', y='Capacity', data=smartphones, hue='OS', size=5)
plt.show()
```

```python
# show base on company 
plt.figure(figsize=(13,8))
sb.swarmplot(x='OS', y='Capacity', data=smartphones, hue='Company', size=8)
plt.show()
```

# Box Plot

```python
#(min , q1, median,q3,max) first=q1(>25%), then median 50%<data<50% ,q3>75% data,
# 1,2,4,5,6,10,12. for Q1 median for( 1,2,3,5), for Q3 median for (5,6,10,12)

sb.boxplot(x='Company', y='Ram', data=smartphones)
plt.show()
```

# Joint Plot

```python
#visualize a bivariate distribution
# kind: means use  scatter formain chart
# we can see histogram capicity and ram on chart
sb.jointplot(x='Capacity', y='Ram', data=smartphones,kind='scatter')
plt.show()
```

# Pair Plot

```python

sb.pairplot(smartphones, hue='Name', palette='hls', plot_kws={'s':80})
plt.show()
```

```python
sb.set_style('darkgrid')
sb.pairplot(smartphones, hue='Name', plot_kws={'s':90})
plt.show()
```
